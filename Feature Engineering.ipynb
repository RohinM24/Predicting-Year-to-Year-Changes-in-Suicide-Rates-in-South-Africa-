```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Feature Engineering Ready Data\\feature_engineering_ready_data.csv")
df.head()
```

```python
predictors = [
    "AlcoholConsumption",
    "HomicideRate",
    "GDP_per_capita",
    "Unemployment, total (% of total labor force) (modeled ILO estimate)_x"
]

target = "SuicideRate"
```

```python
for col in predictors:
    df[col + "_lag1"] = df[col].shift(1)
```

```python
for col in predictors:
    df[col + "_lag2"] = df[col].shift(2)
```

```python
for col in predictors:
    df[col + "_roll3"] = df[col].rolling(window=3).mean()
```

```python
df["Year_scaled"] = df["Year"] - df["Year"].min()
```

```python
df = df.sort_values("Year")
df = df.interpolate(method='linear', limit_direction='both')
```
