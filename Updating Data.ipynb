```python
import pandas as pd
```

```python
df_suicide = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Datasets\\16BBF41_ALL_LATEST.csv")
df_final = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Merged data\\final_model_dataset.csv")
```

```python
df_suicide.head()
df_suicide.columns
```

```python
df_suicide.head()
```

```python
df_sa = df_suicide[
    (df_suicide["GEO_NAME_SHORT"] == "South Africa") &
    (df_suicide["DIM_SEX"] == "TOTAL") &
    (df_suicide["DIM_AGE"] == "TOTAL")
]
```

```python
df_sa_clean = df_sa[["DIM_TIME", 
                      "RATE_PER_100000_N", 
                      "RATE_PER_100000_NL", 
                      "RATE_PER_100000_NU"]].copy()

df_sa_clean.columns = ["Year", "SuicideRate", "SuicideRate_Lower", "SuicideRate_Upper"]
```

```python
df_merged = df_final.merge(df_sa_clean, on="Year", how="left")
```

```python
df_merged
```

```python
df_merged.to_csv("final_model_dataset_UPDATED.csv", index=False)
```

```python
df_filtered = df_merged[df_merged["Year"] >= 2000].copy()
```

```python
df_filtered.reset_index(drop=True, inplace=True)
```
