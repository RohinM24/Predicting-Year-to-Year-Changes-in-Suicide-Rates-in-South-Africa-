```python
import pandas as pd
```

```python
suicide = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Cleaned Data\\cleaned_suicide_rate.csv")
alcohol = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Cleaned Data\\cleaned_alcohol_consumption.csv")
homicide = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Cleaned Data\\cleaned_homicide_rate.csv")
gdp = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Cleaned Data\\cleaned_gdp_per_capita.csv")

# Optional predictors
unemployment_health = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Cleaned Data\\cleaned_unemployment_health.csv")
unemployment_meta = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Cleaned Data\\unemploymentvshealthmetadatacleaned.csv")
```

```python
suicide.head()
alcohol.head()
homicide.head()
gdp.head()
```

```python
suicide = suicide.rename(columns={"Time": "Year"})
```

```python
suicide["Year"] = suicide["Year"].astype(int)
alcohol["Year"] = alcohol["Year"].astype(int)
homicide["Year"] = homicide["Year"].astype(int)
gdp["Year"] = gdp["Year"].astype(int)

# Optional ones
unemployment_health["Year"] = unemployment_health["Year"].astype(int)
unemployment_meta["Year"] = unemployment_meta["Year"].astype(int)
```

```python
merged = suicide.copy()
```

```python
merged = merged.merge(alcohol, on="Year", how="outer")
```

```python
merged = merged.merge(homicide, on="Year", how="outer")
```

```python
merged = merged.merge(gdp, on="Year", how="outer")
```

```python
merged = merged.merge(unemployment_health, on="Year", how="outer")
merged = merged.merge(unemployment_meta, on="Year", how="outer")
```

```python
merged = merged.sort_values("Year").reset_index(drop=True)
merged.head(15)
```
