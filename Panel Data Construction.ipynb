```python
def clean_country_names(df):
    df['Country'] = (
        df['Country']
        .astype(str)
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
    )
    return df

df_suicide = clean_country_names(df_suicide)
df_alcohol = clean_country_names(df_alcohol)
df_homicide = clean_country_names(df_homicide)
df_gdp = clean_country_names(df_gdp)
```

```python
for df in [df_suicide, df_alcohol, df_homicide, df_gdp]:
    df['Year'] = pd.to_numeric(df['Year'], errors='coerce').astype('Int64')
```

```python
df_suicide = df_suicide.dropna(subset=['Year'])
df_alcohol = df_alcohol.dropna(subset=['Year'])
df_homicide = df_homicide.dropna(subset=['Year'])
df_gdp = df_gdp.dropna(subset=['Year'])
```

```python
countries_common = (
    set(df_suicide['Country'])
    & set(df_alcohol['Country'])
    & set(df_homicide['Country'])
    & set(df_gdp['Country'])
)

len(countries_common), list(sorted(countries_common))[:10]
```

```python
df_suicide = df_suicide[df_suicide['Country'].isin(countries_common)]
df_alcohol = df_alcohol[df_alcohol['Country'].isin(countries_common)]
df_homicide = df_homicide[df_homicide['Country'].isin(countries_common)]
df_gdp = df_gdp[df_gdp['Country'].isin(countries_common)]
```

```python
years_common = (
    set(df_suicide['Year'])
    & set(df_alcohol['Year'])
    & set(df_homicide['Year'])
    & set(df_gdp['Year'])
)

min(years_common), max(years_common)
```

```python
df_suicide = df_suicide[df_suicide['Year'].isin(years_common)]
df_alcohol = df_alcohol[df_alcohol['Year'].isin(years_common)]
df_homicide = df_homicide[df_homicide['Year'].isin(years_common)]
df_gdp = df_gdp[df_gdp['Year'].isin(years_common)]
```

```python
df_panel = (
    df_suicide
    .merge(df_alcohol, on=['Country', 'Year'], how='inner')
    .merge(df_homicide, on=['Country', 'Year'], how='inner')
    .merge(df_gdp, on=['Country', 'Year'], how='inner')
)
```

```python
df_panel.head()
df_panel.shape
```

```python
df_panel.head()
```

```python
df_panel = df_panel.sort_values(['Country', 'Year']).reset_index(drop=True)
```

```python
df_panel = df_panel.rename(columns={
    'RATE_PER_100000_N_x': 'SuicideRate',
    'RATE_PER_CAPITA_N': 'AlcoholRate',
    'RATE_PER_100000_N_y': 'HomicideRate'
})
```

```python
df_panel.head()
```

```python
df_panel['SuicideRate_diff'] = (
    df_panel
    .groupby('Country')['SuicideRate']
    .diff()
)
```

```python
df_panel = df_panel.dropna().reset_index(drop=True)
```

```python
for col in ['AlcoholRate', 'HomicideRate', 'GDP_per_capita']:
    df_panel[f'{col}_roll3'] = (
        df_panel
        .groupby('Country')[col]
        .rolling(window=3)
        .mean()
        .reset_index(level=0, drop=True)
    )
```

```python
df_panel = df_panel.dropna().reset_index(drop=True)
```

```python
df_panel.groupby('Country').size()
```

```python
df_train = df_panel[df_panel['Country'] != 'South Africa']
df_sa_panel = df_panel[df_panel['Country'] == 'South Africa'].copy()
```

```python
features = [
    'AlcoholRate_roll3',
    'HomicideRate_roll3',
    'GDP_per_capita_roll3'
]

X_train = df_train[features]
y_train = df_train['SuicideRate_diff']

X_sa = df_sa_panel[features]
y_sa = df_sa_panel['SuicideRate_diff']
```

```python
from sklearn.linear_model import Ridge

model = Ridge(alpha=100)
model.fit(X_train, y_train)
```

```python
sa_preds = model.predict(X_sa)

df_sa_panel['Predicted_diff'] = sa_preds
df_sa_panel[['Year', 'SuicideRate_diff', 'Predicted_diff']].head()
```

```python
df_sa_panel['Predicted_SuicideRate'] = (
    df_sa_panel['SuicideRate'].shift(1) + df_sa_panel['Predicted_diff']
)
```
