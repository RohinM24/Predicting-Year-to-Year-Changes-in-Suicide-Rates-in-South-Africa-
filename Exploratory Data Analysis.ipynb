```python
df = pd.read_csv("C:\\Users\\rohin\\Sucide Project\\Merged data\\final_model_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12, 6))  # optional: control the size

sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")

plt.savefig("missing_values_heatmap.png", dpi=300, bbox_inches="tight")  # <-- save the image
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["SuicideRate"], marker="o")
plt.title("Suicide Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Suicide Rate (per 100k)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["AlcoholConsumption"], color="orange")
plt.title("Alcohol Consumption Over Time")
plt.xlabel("Year")
plt.ylabel("Litres per adult")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["HomicideRate"], color="red")
plt.title("Homicide Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Homicide Rate per 100k")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"],df["GDP_per_capita"],color= "green")
plt.title("GDP Per Capita Over Time")
plt.xlabel("Year")
plt.ylabel("USD")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")

# Save the heatmap to a file
plt.savefig("correlation_heatmap.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
sns.pairplot(df.dropna(), 
             vars=["SuicideRate", "AlcoholConsumption", "HomicideRate", "GDP_per_capita"])
plt.show()
```
